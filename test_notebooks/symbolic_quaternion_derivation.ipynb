# Définition de la molécule
Soit une molécule d'eau à trois sites, les positions des sites sont dans le repère de la molécule:
\begin{align}
    \vec{r_O} &= (0,0,0) \\
    \vec{r_{H1}} &= r_{OH}(0,s,c)\\
    \vec{r_{H2}} &= r_{OH}(0,-s,c)
\end{align}
où $s = sin(\theta/2)$, $c = \cos(\theta/2)$ et les constantes $r_{OH}$ et $\theta$ sont données par le modèle (ici SPC/E)

# Définition des positions relatives entre deux molécules
Soit deux molécules d'eau à des positions $\vec{r_1}$ et $\vec{r_2}$, définies par des positions angulaires quaternioniques $q_1$, $q_2$. On ramnène le premier quaternion comme origine et définissons la position relative de la deuxième molécule:
\begin{align}
    \vec{r} &= \vec{r_2} - \vec{r_1}\\
    q &= q_1^{-1} q_2
\end{align}

Les positions des sites de la deuxième molécule dans ce repère sont alors:
\begin{align}
    \vec{r_O'} &= \vec{r} + q \otimes q_{\vec{r_O}} \otimes q'\\
    \vec{r_{H1}'} &= \vec{r} + q \otimes q_{\vec{r_{H1}}} \otimes q'\\
    \vec{r_{H2}'} &= \vec{r} + q \otimes q_{\vec{r_{H2}}} \otimes q'\\
\end{align}

où $q_{\vec{r_i}}$ est le quaternion pur de $\vec{r_i}$.
