```
ro = Quaternion[0,0,0,0]
rh1 = Quaternion[0, roh * Sin[t/2], 0, roh * Cos[t/2]]
rh2 = Quaternion[0, roh * -Sin[t/2], 0, roh * Cos[t/2]]
rm = Quaternion[0, 0, 0, rom]
```

```
q = Quaternion[a, b, c, d]
qp = Conjugate[q]
```

```
r = Quaternion[0, x, y, z]
```

```
row = r
rh1w = r + q**rh1**qp//Simplify
rh2w = r + q**rh2**qp//Simplify
rmw = r + q**rm**qp//Simplify
```

```
rov1 = {x1,y1,z1}

rh1v1 = {x1 + 2(a1 c1 + b1 d1) roh Cos[t/2] + (a1^2 + b1^2 - c1^2 - d1^2) roh Sin[t/2],
 y1 - 2(a1 b1 - c1 d1) roh Cos[t/2] + 2(b1 c1 + a1 d1) roh Sin[t/2], z1 + (a1^2 - b1^2 - c1^2 + d1^2) roh Cos[t/2] - 2(a1 c1 - b1 d1) roh Sin[t/2]}

rh2v1 = {x1 + 2(a1 c1 + b1 d1) roh Cos[t/2] - (a1^2 + b1^2 - c1^2 - d1^2) roh Sin[t/2],
 y1 - 2(a1 b1 - c1 d1) roh Cos[t/2] - 2(b1 c1 + a1 d1) roh Sin[t/2], z1 + (a1^2 - b1^2 - c1^2 + d1^2) roh Cos[t/2] + 2(a1 c1 - b1 d1) roh Sin[t/2]}

rmv1 = {2 a1 c1 rom + 2 b1 d1 rom + x1, -2 a1 b1 rom + 2 c1 d1 rom + y1, a1^2 rom - b1^2 rom - c1^2 rom + d1^2 rom + z1}

rov2 = {x2,y2,z2}

rh1v2 = {x2 + 2(a2 c2 + b2 d2) roh Cos[t/2] + (a2^2 + b2^2 - c2^2 - d2^2) roh Sin[t/2],
 y2 - 2(a2 b2 - c2 d2) roh Cos[t/2] + 2(b2 c2 + a2 d2) roh Sin[t/2], z2 + (a2^2 - b2^2 - c2^2 + d2^2) roh Cos[t/2] - 2(a2 c2 - b2 d2) roh Sin[t/2]}

rh2v2 = {x2 + 2(a2 c2 + b2 d2) roh Cos[t/2] - (a2^2 + b2^2 - c2^2 - d2^2) roh Sin[t/2],
 y2 - 2(a2 b2 - c2 d2) roh Cos[t/2] - 2(b2 c2 + a2 d2) roh Sin[t/2], z2 + (a2^2 - b2^2 - c2^2 + d2^2) roh Cos[t/2] + 2(a2 c2 - b2 d2) roh Sin[t/2]}

rmv2 = {2 a2 c2 rom + 2 b2 d2 rom + x2, -2 a2 b2 rom + 2 c2 d2 rom + y2, a2^2 rom - b2^2 rom - c2^2 rom + d2^2 rom + z2}
```

```
r1i := rov1
r2i := rh1v1
r3i := rh2v1
r4i := rmv1
r1j := rov2
r2j := rh1v2
r3j := rh2v2
r4j := rmv2
ri = {r1i, r2i, r3i, r4i};
rj = {r1j, r2j, r3j, r4j};
q = {0, qh, qh, qm};

U = 1/(4 Pi 0) Sum[
  (q[[a]] * q[[b]]) / Norm[ri[[a]] - rj[[b]]],
  {a, 1, 4}, {b, 1, 4}
]//Simplify
```

```
Grad[U, {x1,y1,z1}]
```
